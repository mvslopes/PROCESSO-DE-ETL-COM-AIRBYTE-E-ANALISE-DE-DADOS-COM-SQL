##**SPRINT 02 - CRIAR E INICIAR UM CONTAINER DOCKER COM BANCO DE DADOS POSTGRESQL**

###**01- Objetivo da Sprint 02:**

Na **Sprint 02**, o objetivo foi a criação e inicialização de um container Docker contendo uma **instância do banco de dados PostgreSQL**. Ao utilizar esse comando, efetivamente criamos um ambiente isolado que hospeda o **PostgreSQL**, permitindo a configuração de um banco de dados específico, juntamente com suas configurações de usuário.

A principal vantagem desse processo é que o banco de dados **PostgreSQL** se torna acessível localmente por meio da uma determinada porta do sistema hospedeiro, simplificando as tarefas de desenvolvimento, testes e outras atividades relacionadas a bancos de dados.

Esse ambiente isolado garante que qualquer alteração ou operação realizada no banco de dados **PostgreSQL** não afetará o sistema hospedeiro, tornando-o uma escolha ideal para ambientes de desenvolvimento, experimentação e testes.

###**03- Criando e inicializando um container Docker com uma instância do banco de dados PostgreSQL**

**3.1- Para Criar o banco de dados com o docker foi utilizado o codigo abaixo:**

```python
docker run --name projetoairbyte -p xxxx:yyyy -e POSTGRES_USER=mvslopes -e POSTGRES_PASSWORD=xxxxxx -e POSTGRES_DB=projeto03 -d postgres
```

**--name projetoairbyte:**
* Isso atribui um nome ao container, que neste caso é "projetoairbyte". O nome é usado para identificar e referenciar o container posteriormente.



**-p xxxx:yyyy:**

* Especifica a porta que o container irá expor. O formato é **[ porta do sistema host ] : [ porta do container ]**. Neste caso, o container **PostgreSQL** expõe a **porta xxxx**, que é a porta padrão do **PostgreSQL**, para a **porta yyyy** do sistema host.

**-e POSTGRES_USER=mvslopes:**
* Define uma variável de ambiente no container. Neste caso, está definindo o usuário do PostgreSQL como "mvslopes".

**-e POSTGRES_PASSWORD=#######:** Define outra variável de ambiente no container, especificando a senha para o usuário do PostgreSQL, que é "empresa88".

**-e POSTGRES_DB=projeto03:** Define mais uma variável de ambiente, que especifica o nome do banco de dados PostgreSQL, neste caso, **"airbyte"**.

**-d postgres:** Especifica a imagem do container a ser utilizada, que é a imagem oficial do PostgreSQL. O Docker irá baixar essa imagem e iniciar o contêiner com as configurações especificadas.

Em resumo, esse comando cria e inicia um container Docker baseado na imagem oficial do **PostgreSQL**, configurando um usuário, senha e nome do banco de dados para o **PostgreSQL** no container e expondo a porta xxxx para permitir a conexão com o banco de dados. O nome do contêiner é definido como **"projetoairbyte"**.

###**4- Conectando o contêiner ao banco de dados do PgAdmin via localhost**

* Conectar o **container** ao banco de dados do **PgAdmin** via localhost significa estabelecer uma conexão entre o **PgAdmin**, que é uma ferramenta de gerenciamento e administração de bancos de dados **PostgreSQL**, e o container do banco de dados PostgreSQL que está em execução na mesma máquina **(localhost).**

* Essa conexão permite que você acesse e gerencie o banco de dados PostgreSQL hospedado no container diretamente pelo **PgAdmin**. É uma forma de administrar o banco de dados de maneira mais amigável e visual, **executando consultas, realizando tarefas de manutenção, visualizando dados e executando outras operações por meio da interface do PgAdmin, em vez de usar comandos SQL brutos no terminal.**

* Para conectar o container ao **PgAdmin** via localhost, geralmente você precisa fornecer informações como o nome do servidor, porta, nome de usuário e senha, permitindo ao PgAdmin estabelecer a conexão com o banco de dados PostgreSQL no container e interagir com ele. Isso simplifica a administração do banco de dados e facilita a execução de tarefas de gerenciamento.
