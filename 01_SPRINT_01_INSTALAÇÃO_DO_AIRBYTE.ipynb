##**SPRINT 01- INSTALAÇÃO DO AIRBYTE**

###**01- Objetivo da Sprint 01**

**Na Sprint 01**, o objetivo foi a instalação e configuração do **Airbyte**. Para isso, foi utilizado a documentação oficial, disponível em **https://airbyte.com/**. Ao acessar o site do **Airbyte**, o próximo passo foi clicar em **"Deploy Airbyte Source"** para realizar a instalação.

* O Airbyte é uma plataforma de código aberto usada para a ingestão e sincronização de dados de várias fontes para um data warehouse, data lake ou outro destino de dados.

O principal objetivo da **instalação do Airbyte** é permitir que as organizações coletem, consolidem e movam dados de maneira eficiente. Isso é essencial para vários cenários, incluindo melhorar a tomada de decisões baseadas em dados e otimizar as operações das organizações.

**Número 01: Sources**

* As **"sources"** ou **Fontes** no Airbyte são componentes específicos que sofre configuração para se conectar a uma fonte de dados específica. Esses **componentes** são responsáveis por extrair dados da fonte e torná-los disponíveis para serem **transformados, enriquecidos e carregados em destinos de dados**. O Airbyte oferece uma variedade de **"sources"** pré-construídas para uma ampla gama de sistemas, o que facilita a configuração e a coleta de dados de diversas fontes.

* Essas **"sources"** incluem adaptadores para bancos de dados relacionais, bancos de dados NoSQL, serviços da web, aplicativos de terceiros e muito mais. Além disso, você pode criar suas próprias "sources" personalizadas se precisar integrar dados de uma fonte não suportada nativamente.


**Número 02: Destinations**

* No Airbyte, **"destinations"** (destinos) são os sistemas ou locais para onde os dados coletados e transformados são direcionados e armazenados. Os **"destinations"** são a parte final do processo de integração de dados, onde as informações são entregues e organizadas de acordo com as necessidades do usuário.

* **Os "destinations"** podem ser vários tipos de sistemas de armazenamento de dados, **bancos de dados, data warehouses, data lakes, serviços de nuvem, aplicativos de análise, entre outros**. O Airbyte oferece uma variedade de **"destinations"** pré-construídos e suporte para conexões personalizadas, permitindo que escolha o local mais apropriado para os dados, dependendo das necessidades do seu projeto.

**Número 03: Connections**

* No Airbyte, "connections" (conexões) são configurações que permitem estabelecer uma ligação segura entre uma **"source" (fonte)** e um **"destination" (destino)**. As conexões são usadas para autenticar e autorizar o acesso aos sistemas de origem e destino de dados. **Elas fornecem informações, como credenciais de autenticação, endereços URL e outras configurações específicas para garantir que os dados possam ser transferidos com segurança entre esses sistemas.**

* **O Airbyte** oferece conexões pré-configuradas para muitos sistemas populares, tornando o processo de configuração mais fácil e rápido. Além disso, você pode criar conexões personalizadas, caso deseje integrar sistemas que não estejam diretamente suportados pelas conexões pré-configuradas.
