##**SPRINT 03 - CONFIGURANDO "DESTINATIONS," "SOURCE," E "CONNECTIONS" NO AIRBYTE.**

###**01- Objetivo da Sprint 03**

O **Airbyte** é uma plataforma de código aberto para integração de dados que permite coletar, transformar e carregar **(ETL)** dados de várias fontes para vários destinos. As etapas de configuração no Airbyte envolvem a criação de fontes (sources), conexões (connections) e destinos (destinations). Aqui está uma explicação das etapas:

**O Objetivo dessa sprint foi configurar as etapas abaixo:**

* **Sources (Fontes):** As fontes são as origens dos dados que desejamos coletar e integrar no **Airbyte**. O primeiro passo é configurar uma fonte, que pode ser uma conexão com um sistema de banco de dados, uma API, um arquivo CSV, entre outros. Foi definido as configurações específicas da fonte, como as **credenciais de acesso**, **o local dos dados** e **os parâmetros de conexão.**
Depois de configurar a fonte, você pode testá-la para garantir que a conexão seja bem-sucedida e que os dados possam ser acessados.

* **Connections (Conexões):** As conexões são os canais que ligam as fontes aos destinos. Elas representam os fluxos de dados que você deseja estabelecer.
Ao criar uma conexão, você seleciona uma fonte **(source)** previamente configurada e um destino (destination) onde os dados serão carregados. Você pode **definir transformações específicas para as conexões, como mapear campos, filtrar dados ou aplicar lógica personalizada.** É importante garantir que a conexão esteja corretamente configurada para que os dados sejam transferidos de forma adequada.

* **Destinations (Destinos):** Os destinos são os locais para onde os dados serão carregados a partir das fontes. Você configura um destino para cada conexão, definindo as configurações específicas, como credenciais de acesso e localização dos dados no destino. **O Airbyte suporta vários tipos de destinos, como bancos de dados, armazenamento em nuvem, data warehouses e mais.** Após configurar o destino, você pode testar a conexão para garantir que os dados sejam carregados com sucesso no destino pretendido.


* Uma vez configurado, o Airbyte permite agendar e automatizar a execução das conexões para garantir que os dados sejam atualizados regularmente, facilitando a integração contínua de informações de várias fontes para fins analíticos, relatórios e muito mais.

**4.5- Criar uma fonte no Airbyte refere-se ao processo de configurar um local onde os dados serão extraídos**.

* **Source Name (Nome da Fonte):** Nome que daremos à fonte de dados . É uma identificação amigável que ajuda a nomear e organizar as fontes de dados em sua configuração no Airbyte. O "Source Name" pode ser um nome descritivo que reflete a origem ou o propósito da fonte de dados.

* **Dataset Name (Nome do Conjunto de Dados):**o "Dataset" representa O nome da tabela final para replicar este arquivo. Éo nome da tabela final  que queremos levar para o Data Warehouse que é o nosso destino.

* **Provedor de Armazenamento (Storage Provider):** é à origem dos dados. Por exemplo, se os dados estão armazenados localmente em seu dispositivo, o provedor de armazenamento seria o **"Local"**. Se os dados estão em um serviço em nuvem, como o Amazon S3 ou o Google Cloud Storage, você selecionaria o provedor apropriado. A escolha do provedor determina como a plataforma acessará os dados.

* **HTTPS:** Este campo refere-se ao protocolo de segurança usado para a transferência de dados. A opção **HTTPS (Hypertext Transfer Protocol Secure)** é a versão segura do HTTP e é amplamente utilizada para garantir a segurança das comunicações online.

* **URL (Uniform Resource Locator)** é o endereço da web que aponta para os dados que você deseja importar. Você deve fornecer o URL correto que leva aos seus dados, e a plataforma Airbyte usará esse URL para acessar e importar os dados relevantes.

Essas configurações são fundamentais para garantir que a plataforma AI/Byte possa acessar e processar seus dados de forma adequada. Certifique-se de preencher essas informações com precisão para uma integração bem-sucedida dos dados na plataforma.

**5.7-Comentando as configurações:**

* **Connection name:** Qual o nome da conexão, escolhemos o nome **carrega_produtos_cabelos**  Pode ser o nome que você quiser.

* **Replication frequency:** A frequência de replicação, de quanto em quanto tempo nós vamos executar esse conector? É basicamente para agendar a carga de dados. Para o nosso projeto foi usado a opção **'manual'** .

* **Destination Namespace:** Como definimos o namespace? Queremos um espelho da estrutura da origem? Pegamos a estrutura da origem e replicamos no destino? Sim, é o que o projeto estabeleceu. **Observação importante:** Temos também outras opções. Nessa fase, podemos personalizar o que teremos no destino. Isso é típico de ferramentas ETL. A **Source** seria Extração, o **Destination** seria o Load, que é a carga, e a **Connections** seria a transformação. Esse tópico foi configurado como **Mirror source structure.**

* **Destination Stream PrefixPois:** Estabelece que podemos escolher um prefixo para as tabelas que irão para o destino. Para a tabela dos projetos não colocamos préfixos.

* **Activate the streams you want to syncgunt:** Nesse tópico temos duas opções. Primeira opção, **"Full Refresh | Overwiter"** Nesse caso, a operação vai sobrescrever o que existe no destino. A segunda Opção **"Full Refresh | Append"** se a tabela já tem algum dado, a operação vai adicionar algo à tabela já existente. Então, no nosso projeto, não temos nada no banco de dados, então o projeto vai usar **"Full Refresh | Overwiter"**, que é a carga inicial. Se o executar uma segunda carga, pdemos escolher sobrescrever. Há uma terceira opção que não está aparecendo, seria o **'refresh incremental'**. Para configurar o **'refresh incremental'**, você precisa detectar uma chave primária. Como não temos nenhuma definição de chave aqui nos arquivos, ele não consegue usar. Não aparece na lista.

* **Normalization & Transformationqual:** Por fim o sistema pergunta qual a  normalização. Para o nosso projeto, os dados são em formato tabular. Mas é possível customizar isso também.

**Observação 01**:

* As tabelas que estão marcadas de vermelho são as tabelas **"airbyte_raw"** e  podem ser consideradas como tabelas de controle usadas para gerenciar o processo de ETL **(Extração, Transformação e Carga)** e manter o controle sobre os dados brutos antes que sejam transformados e carregados nas tabelas de destino.

* Essas tabelas de controle ajudam a gerenciar e monitorar o **fluxo de dados** durante o processo de integração. Elas desempenham várias funções, incluindo:

*  **Armazenamento de Dados Brutos:** As tabelas **"airbyte_raw"** armazenam os dados brutos extraídos da fonte antes que qualquer transformação seja aplicada. Isso permite que mantenha um registro dos dados originais.

* **Registro e Auditoria:** Essas tabelas registram informações sobre o processo de ETL, como **timestamps**, detalhes de fonte, status de processamento e outros metadados relacionados aos dados.

* **Transformação Temporária:** As tabelas **"airbyte_raw"** podem ser usadas para aplicar temporariamente transformações nos dados antes de serem carregados nas tabelas de destino. Isso pode incluir limpeza, validação e mapeamento de campos.

* **Integridade e Recuperação de Dados:** Elas também desempenham um papel na garantia de integridade dos dados e na recuperação em caso de falhas durante o processo de ETL.

Portanto, essas tabelas de controle são essenciais para manter a qualidade dos dados, rastrear o processo de ETL e fornecer uma camada intermediária onde as transformações podem ser aplicadas. **Elas não são tabelas de destino finais, mas são uma parte crítica do fluxo de integração de dados do Airbyte. e eslas podem influenciar no tamanho do banco de dados**
