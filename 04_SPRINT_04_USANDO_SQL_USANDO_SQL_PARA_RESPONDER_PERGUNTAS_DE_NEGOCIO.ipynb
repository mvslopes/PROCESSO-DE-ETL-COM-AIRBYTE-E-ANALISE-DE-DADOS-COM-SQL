###**01- Objetivo da Sprint 04**

**1.1- Conhecendo a área de Negócio**

Os dados empregados no projeto referem-se à produtividade de oito linhas de separação de cosméticos do centro de distribuição **New Cosmetic**, abrangendo o período de **janeiro de 2023 a outubro de 2023**.

Vale ressaltar que esse centro de distribuição opera com linhas de separação de pedidos que fazem uso da tecnologia Pick to Light, também conhecida como "separação por luzes". Este centro de distribuição desempenha um papel crucial na estratégia de distribuiçã de um grande fabricante de para produtos de cabelos e pele.

O objetivo principal da análise dos dados de produtividade (separação de pedidos) usando SQL é encontrar insights para que os gestores possam tomar decisões baseadas em dados, aprimorando a atividade de picking e a eficiência operacional.

---

**1.2- Plano de Anáilse**



1. **Definição do objetivo:** O objetivo principal é analisar a produtividade, separação de peças (unidades de produtos/cosméticos) por hora no **Centro de Distribuição New Cosmetic, no período de janeiro de 2023 a outubro de 2023**. A análise permitirá identificar as linhas que tiveram produtividade abaixo da meta nesse período de separação de peças ao longo de 10 meses. Ao descobrirmos essas linhas com produtividade baixa, as equipes de melhoria contínua do centro de distribuição podem implementar planos de alavancagem de produtividade.

2. **Coleta e extração de dados:** Os dados utilizados foram coletados durante as datas de **02-01-2023 a 31-10-2023**, abrangendo informações como data, hora de separação, linhas de separação, produtos e produtividade das linhas de separação. Isso inclui as duas categorias: **produtos destinados aos cuidados com cabelos e produtos destinados aos cuidados com a pele.**

3. **Escolha de ferramentas:** Foram utilizadas três ferramentas: **O Docker** para criação, implantação e execução dos contêineres do Airbyte e PostgreSQL. **O Airbyte** para coleta, preparação e movimentação de dados entre diferentes fontes e destinos. **O PostgreSQL** para implementação do SQL, permitindo a realização de consultas complexas, atualizações e manipulações de dados de maneira eficiente.

4. **Aplicação das técnicas de análise:** As técnicas de análise foram aplicadas utilizando SQL, buscando entender a eficiência operacional e a atividade de picking no contexto das linhas de separação de pedidos.

5. **Entrega do resultado:** O resultado da análise visa fornecer insights para os gestores, auxiliando-os na tomada de decisões informadas para aprimorar a atividade de picking e a eficiência operacional no centro de distribuição **New Cosmetic**.
